```python
import pandas as pd

data = {'luas' : [6, 8, 10, 14, 18],
        'harga' : [7, 9, 13, 17.5, 18]}

df = pd.DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt

df.plot(kind='scatter', x='luas', y='harga')
plt.title('perbandingan harga dengan luas ')
plt.xlabel('luas tanah (m persegi)')
plt.ylabel('harga tanah (dollar)')
plt.xlim(0, 25)
plt.ylim(0, 25)
plt.grid(True)
plt.show()
```

```python
import numpy as np
x=np.array (df['luas']) #fitur (dependen)
y=np.array (df['harga']) #target (independen)

print (f'x:{x}')
print (f'y:{y}')
```

```python
x = x.reshape(-1, 1)

print (f'x: shape: {x.shape}')
print (f'x: {x}')
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x, y)
```

```python
x_vis =np.array([0, 25]).reshape(-1, 1)
y_vis = model.predict(x_vis)
```

```python
plt.scatter(x, y)
plt.plot(x_vis, y_vis, c='r')
plt.title('perbandingan harga dengan luas')
plt.xlabel('luas tanah (m persegi)')
plt.ylabel('harga tanah (dollar)')
#plt.xlim(0, 25)
#plt.ylim(0, 25)
plt.grid(True)
plt.show()
```

y = a + bx

y= respon variabel
x= explanatory variabel
a= intercept
b; slope

```python
print (f'intercept: {model.intercept_}')
print (f'slope: {model.coef_}')
```

```python
luas_tanah = np.array([12,15,24]).reshape(-1, 1)
print(luas_tanah.shape)
luas_tanah
```

```python
harga_prediksi = model.predict(luas_tanah)
harga_prediksi
```

```python
for luas, hrg in zip(luas_tanah, harga_prediksi):
    print(f'luas: {luas}, harga: {hrg}')
```

## EVALUASI & TESTING DATASET

```python
x_train = np.array([6, 8, 10, 14, 18]).reshape(-1, 1)
y_train = np.array([7, 9, 13, 17.5, 18])

x_test = np.array([9, 9, 11, 16, 12]).reshape(-1, 1)
y_test = np.array([11, 8.5, 15, 18, 11])
```

```python
model=LinearRegression()
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import r2_score

y_pred = model.predict(x_test)
r_squared = r2_score(y_test, y_pred)
print(f'R-squared: {r_squared}')
```
